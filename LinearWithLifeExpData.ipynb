```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\91790\OneDrive\Desktop\DA and ML\Project\Life Expectancy Data.csv")
df
```

```python
df.isnull().sum()
```

```python
missing = df.isnull().sum()
missing = missing[missing > 0]
print(missing)
```

```python
# 1. Remove spaces
df.columns = df.columns.str.strip()

# 2. Select numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns

# 3. Fill missing by group median
df[numeric_cols] = df.groupby("Country")[numeric_cols]\
                       .transform(lambda x: x.fillna(x.median()))

# 4. Fill any remaining globally
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
df.isnull().sum()
```

```python
# Encode Developed / Developing
df['Status'] = df['Status'].map({'Developed': 1, 'Developing': 0})
df = pd.get_dummies(df, columns=['Country'], drop_first=True)
```

```python
x = df.drop('Life expectancy', axis=1)
y = df['Life expectancy']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("MES: ", mse)
print("R2: ", r2)
```
